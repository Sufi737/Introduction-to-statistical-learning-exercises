```python
import pandas as pd
df = pd.read_csv('data/Boston.csv')
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#pairplots of all predictors
import seaborn as sns
sns.pairplot(df,kind='scatter', dropna=True)
```

```python
#finding correlation of predictors with per capita crime rate
corr = df.corr()
corr = corr['crim']
corr
```

```python
#finding suburbs with very high per capita crime rate
val = df['crim'].quantile(0.99)
print(df[df['crim'] > val])
```

```python
#high tax rate suburbs
val = df['tax'].quantile(0.99)
print(df[df['tax'] > val])
```

```python
#high pupil teacher ratio suburbs
val = df['ptratio'].quantile(0.99)
print(df[df['ptratio'] > val])
```

```python
#number of suburbs bound to the Charles river
df[df['chas'] == 1].count()[0]
```

```python
#median of pupil-teacher ratio
df['ptratio'].median()
```
