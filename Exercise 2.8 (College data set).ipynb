```python
"""
College data set consists of 700 records of different universities/colleges
consists of the following attributes:
Private : Public/private indicator 
Apps : Number of applications received 
Accept : Number of applicants accepted 
Enroll : Number of new students enrolled 
Top10perc : New students from top 10% of high school class 
Top25perc : New students from top 25% of high school class 
F.Undergrad : Number of full-time undergraduates 
P.Undergrad : Number of part-time undergraduates 
Outstate : Out-of-state tuition 
Room.Board : Room and board costs 
Books : Estimated book costs 
Personal : Estimated personal spending
PhD : Percent of faculty with Ph.D.’s 
Terminal : Percent of faculty with terminal degree 
S.F.Ratio : Student/faculty ratio 
perc.alumni : Percent of alumni who donate 
Expend : Instructional expenditure per student 
Grad.Rate : Graduation rate 
"""


import pandas as pd
df = pd.read_csv('data/college.csv')
df
```

```python
#setting the name of the colleges as the index of the data set
df.set_index('Unnamed: 0',drop=True,inplace=True)
df.index.names = ['Name']
df
```

```python
#getting a numerical summary of all the columns
df.describe()
```

```python
first_ten = df.iloc[:,1:11]
first_ten
```

```python
"""
this actually did not made sense but the book said to create scatter plot of first ten columns
created a scatter plot by taking pairs of 2
"""
ax1 = first_ten.plot(kind='scatter',x='Apps',y='Accept',color='r')
ax2 = first_ten.plot(kind='scatter',x='Enroll',y='Top10perc',color='b',ax=ax1)
ax3 = first_ten.plot(kind='scatter',x='Top25perc',y='F.Undergrad',color='g',ax=ax2)
ax4 = first_ten.plot(kind='scatter',x='P.Undergrad',y='Outstate',color='orange',ax=ax3)
ax5 = first_ten.plot(kind='scatter',x='Room.Board',y='Books',color='black',ax=ax4)
```

```python
#binning the top10perc variable 
#dividing the universities depending on whether the proportion of top 10% high school students exceeds 50%
df['Elite'] = df['Top10perc'].apply(lambda x: "Yes" if x>50 else "No")
#getting the count of elite universities
df['Elite'].value_counts()
```

```python
#plotting histogram for some of the columns
df.hist(column='Apps',bins=10)
df.hist(column='Accept',bins=10)
df.hist(column='Enroll',bins=10)
df.hist(column='Outstate',bins=10)
```

```python
#finding out the relation between different variables(this is where the fun begins)

#first we should find out what variables are affected when a university is private
private_applicants = df.groupby('Private')['Apps'].mean()
private_applicants
```

```python
#we can see above the number of applicants for non-private universities are higher as compared to private universities
private_accept = df.groupby('Private')['Accept'].mean()
private_accept
```

```python
#we need to find the ratio of acceptance to applications for private and non private universities

private_accept_ratio = (private_accept['Yes']/private_applicants['Yes'])*100
print("Prob of being accepted in private university: "+str(int(private_accept_ratio))+"%")
nonprivate_accept_ratio = (private_accept['No']/private_applicants['No'])*100
print("Prob of being accepted in a non private university: "+str(int(nonprivate_accept_ratio))+"%")
```

```python
#finding the ration of applicants enrolled after being accepted for private/non-private institutions
private_enrolled = df.groupby('Private')['Enroll'].mean()
private_enrolled
```

```python
private_enroll_perc = (private_enrolled['Yes']/private_accept['Yes'])*100
nonprivate_enroll_perc = (private_enrolled['No']/private_accept['No'])*100
print("Percentage of students accepting private institutions: "+str(int(private_enroll_perc))+"%")
print("Percentage of students accepting non-private institutions: "+str(int(nonprivate_enroll_perc))+"%")
```

```python
import matplotlib.pyplot as plt
ax = private_accept.plot(kind='bar',color='b')
private_enrolled.plot(kind='bar',color='r',ax=ax)
ax.legend()
plt.title('Ratio of students enrolled and accepted for private/non-private institutions')
plt.figtext(0.99, -0.1, 'It is clear that number of enrollments for non-private institutions are higher.', horizontalalignment='right')
plt.figtext(0.99, -0.2, 'Also the ratio of students enrolling to non-private institutions is also high(41%) as compared to private institutions(34%)', horizontalalignment='right')

plt.show()
```

```python
#relation between private/non-private colleges and top 10% students new students
top10_dist = df.groupby('Private')['Top10perc'].mean()
top10_dist
```

```python
#we can see the average number of new students who were top 10% of high school are greater in private colleges
#below is the average values for top 25%
top25_dist = df.groupby('Private')['Top25perc'].mean()
top25_dist
```

```python
#again private colleges have a slightly higher number of new students who were top 25% of highschools


#comparing the average total cost of private and non-private colleges

df['Total Cost'] = df['Room.Board']+df['Books']+df['Personal']
cost_dist = df.groupby('Private')['Total Cost'].mean()
cost_dist
```

```python
#for private colleges the average cost is greater
#finding the effect of total cost on students applying for the college

df.plot(x='Total Cost',y='Apps',style='o')
```
